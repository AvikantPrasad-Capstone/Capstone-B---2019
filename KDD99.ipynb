```python
def dummies(df, name):
    dummy = pd.get_dummies(df[name])
    for x in dummy.columns:
        dummy_name = f"{name}-{x}"
        df[dummy_name] = dummy[x]
    df.drop(name, axis=1, inplace=True)
```
